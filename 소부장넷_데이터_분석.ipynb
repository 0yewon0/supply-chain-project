```python
df_sheet1
```

```python
df_sheet1 = df_sheet1.dropna(how='all')

# 현재 인덱스 제거 (기본 인덱스 설정)
df_sheet1.reset_index(drop=True, inplace=True)

# 4번째 행부터 데이터를 선택 (0-based index, 4번째 행은 인덱스 3)
df_sheet1 = df_sheet1.iloc[3:].reset_index(drop=True)

df_sheet1.iloc[0, 0] = '기간'
df_sheet1.set_index(df_sheet1.columns[0], inplace=True)
df_sheet1.columns = df_sheet1.iloc[0]  # 첫 번째 행을 열 이름으로 설정
df_sheet1 = df_sheet1.iloc[1:]
```

```python
df_sheet1.info()
```

```python
df_sheet1.columns = pd.to_datetime(df_sheet1.columns.str.replace(' 현재', ''), format='%Y.%m.%d')

# 데이터프레임을 길게 변환 (melt)하여 날짜가 행으로 오게끔 변환
df_melted = df_sheet1.melt(id_vars=df_sheet1.columns[0],
                            var_name='Date',
                            value_name='Value')

# 데이터프레임을 날짜를 인덱스로 설정
df_melted['Date'] = pd.to_datetime(df_melted['Date'])
df_melted.set_index('Date', inplace=True)

# 원래의 열 이름을 행으로 변환
df_melted.rename(columns={df_sheet1.columns[0]: 'Category'}, inplace=True)
```

```python
# '부채 총계'와 '자본 총계' 행을 추출
debt_total_row = df_sheet1.loc['부채총계']
equity_total_row = df_sheet1.loc['자본총계']

# 부채 비율 계산
debt_ratio = debt_total_row / equity_total_row

# 부채 비율 행을 데이터프레임으로 변환
debt_ratio_row = pd.DataFrame([debt_ratio], index=['부채비율'])

# 원본 데이터프레임에 부채 비율 행 추가
df_sheet1 = pd.concat([df_sheet1, debt_ratio_row])
```

```python
# 열 이름에서 '수출금액'이 포함된 열만 선택
def is_export_amount(col_name):
    return '수출금액' in col_name

# 수출금액 관련 열만 필터링
filtered_data = data.loc[:, [col for col in data.columns if is_export_amount(col)]]

# 열 이름에서 날짜 부분만 추출
def extract_date_from_column(col_name):
    match = re.search(r'\((\d{6})\)', col_name)
    if match:
        return match.group(1)
    return None
```

```python
from sklearn.linear_model import LinearRegression

# 12개월 후 예측을 위한 기간 설정
future_months = 12

# 그래프를 그리기 위한 색상 팔레트 설정
colors = plt.get_cmap('tab20').colors

plt.figure(figsize=(12, 8))

for idx, column in enumerate(filtered_data.columns):
    single_series = filtered_data[column]

    # 인덱스가 DatetimeIndex인지 확인하고 변환
    if not isinstance(single_series.index, pd.DatetimeIndex):
        single_series.index = pd.to_datetime(single_series.index)

    # 시간 인덱스를 숫자형으로 변환
    x = np.arange(len(single_series)).reshape(-1, 1)
    y = single_series.values

    # 모델 적합
    model = LinearRegression()
    model.fit(x, y)

    # 12개월 후 예측
    future_x = np.arange(len(single_series) + future_months).reshape(-1, 1)
    future_y = model.predict(future_x)

    # 예측 날짜 생성
    future_dates = pd.date_range(start=single_series.index[-1] + pd.DateOffset(months=1), periods=future_months, freq='M')

    # 결과 시각화
    plt.plot(single_series.index, single_series, marker='o', linestyle='-', color=colors[idx], label=f'{column} 원본')
    plt.plot(future_dates, future_y[-future_months:], linestyle='--', color=colors[idx], label=f'{column} 예측')

plt.title('수출금액 시계열 데이터 및 선형 회귀 예측')
plt.xlabel('날짜')
plt.ylabel('수출금액')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='분류', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

plt.figure(figsize=(12, 8))

for idx, column in enumerate(filtered_data.columns):
    single_series = filtered_data[column]

    # 인덱스가 DatetimeIndex인지 확인하고 변환
    if not isinstance(single_series.index, pd.DatetimeIndex):
        single_series.index = pd.to_datetime(single_series.index)

    # 모델 적합
    model = ExponentialSmoothing(single_series, trend='add', seasonal=None, seasonal_periods=12)
    fit = model.fit()

    # 12개월 후 예측
    forecast = fit.forecast(steps=future_months)

    # 예측 날짜 생성
    future_dates = pd.date_range(start=single_series.index[-1] + pd.DateOffset(months=1), periods=future_months, freq='M')

    # 결과 시각화
    plt.plot(single_series.index, single_series, marker='o', linestyle='-', color=colors[idx], label=f'{column} 원본')
    plt.plot(future_dates, forecast, linestyle='--', color=colors[idx], label=f'{column} 예측')

plt.title('수출금액 시계열 데이터 및 지수 평활법 예측')
plt.xlabel('날짜')
plt.ylabel('수출금액')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='분류', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA

plt.figure(figsize=(12, 8))

for idx, column in enumerate(filtered_data.columns):
    single_series = filtered_data[column]

    # 인덱스가 Datetime Index인지 확인하고 변환
    if not isinstance(single_series.index, pd.DatetimeIndex):
        single_series.index = pd.to_datetime(single_series.index)

    # 모델 적합
    model = ARIMA(single_series, order=(2, 1, 0))  # (p, d, q) 값은 데이터에 따라 조정
    fit = model.fit()

    # 12개월 후 예측
    forecast = fit.forecast(steps=future_months)

    # 예측 날짜 생성
    future_dates = pd.date_range(start=single_series.index[-1] + pd.DateOffset(months=1), periods=future_months, freq='M')

    # 결과 시각화
    plt.plot(single_series.index, single_series, marker='o', linestyle='-', color=colors[idx], label=f'{column} 원본')
    plt.plot(future_dates, forecast, linestyle='--', color=colors[idx], label=f'{column} 예측')

plt.title('수출금액 시계열 데이터 및 ARIMA 예측')
plt.xlabel('날짜')
plt.ylabel('수출금액')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='분류', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
